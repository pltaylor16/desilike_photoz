```python
from desy3_observables import DESY3Observable
from desy3_theories import DESY3Theory
```

```python
from desilike.theories import Cosmoprimo
cosmo = Cosmoprimo(engine='camb')

#set some cosmological parameters and one of the weak lensing nuissance paramters
cosmo.init.params = {'Omega_m': {'prior': {'limits': [0.1, 0.9]},
                          'ref': {'dist': 'norm', 'loc': 0.3, 'scale': 0.002},
                          'latex': '\Omega_m'},
                     'A_s': {'prior': {'limits': [1.5e-9, 3e-9]},
                          'ref': {'dist': 'norm', 'loc': 2e-9, 'scale': 5e-10},
                          'latex': 'A_s'},
                    'intrinsic_alignment_parameters_a1':{'prior': {'limits': [-0.1, 0.1]},
                          'latex': 'a_1'}}
```

```python
ini_file_name = 'des-y3.ini'
ini_file_dir = '/srv/scratch1/taylor.4264/test/kp7/desilike_des_module/ini_files'
cosmosis_dir = '/srv/scratch1/taylor.4264/codes/cosmosis/cosmosis-standard-library'

theory = DESY3Theory(cosmo=cosmo, ini_file_dir=ini_file_dir, ini_file_name=ini_file_name, cosmosis_dir=cosmosis_dir)
observable = DESY3Observable(ini_file_dir=ini_file_dir, ini_file_name=ini_file_name, cosmosis_dir=cosmosis_dir, theory=theory) 
```

```python
from desilike.likelihoods import ObservablesGaussianLikelihood

likelihood = ObservablesGaussianLikelihood(observables=[observable])
```

```python
#run the likelihood while changing an IA parameter and Omega_m
likelihood(Omega_m = 0.2, intrinsic_alignment_parameters_a1 = 0.05)
```
